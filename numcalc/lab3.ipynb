```python
import numpy as np
import matplotlib.pyplot as plt
h = 0.001
```

```python
def F(x):
  return (x - 3) * np.cos(x) - 1

def d(foo, x):
  return (foo(x + h) - foo(x)) / h
```

```python
x = np.arange(-2 * np.pi, 2 * np.pi, h)
y = F(x)
plt.plot(x, y)
plt.plot(x, x * 0)
height = [-2, 2]
plt.vlines([-5, -4, -2, -1, 5, 5.5], -2, 2, colors=['green', 'green', 'red', 'red', 'blue', 'blue'], linestyles='dashed')
plt.show()
print("Выделены отрезки: -5..-4, -2..-1, 5..5,5")
print("Далее рассмотрю -5..-4")
```

```python
x = np.arange(-5, -4, h)
plt.plot(x, F(x))
plt.hlines(0, -5, -4)
plt.show()
print("График F(x) на отрезке -5..-4")
```

```python
tau = -1/5
plt.plot(x, tau * d(F, x))
plt.hlines([-2, 0], xmax=-5, xmin = -4)
plt.show()
print("Подберём f(x) = x + tau * F(x)")
print("Для этого подберём такое tau, чтобы зажать tau * F'(x) между 0 и -2")
print("tau = -1/5")
```

```python
def f(x):
  return x + tau * F(x)
```

```python
def iter(eps):
  x_cur = -5
  k = 0
  while np.abs(x_cur - f(x_cur)) > eps:
    k += 1
    x_cur = f(x_cur)
  return k, x_cur

eps = 0.001
k, x_sol = iter(eps)
print("Метод релаксации")
print("Eps =", eps)
print("Число итераций", k)
print("Уточнённое решение", x_sol)
```

```python
x = np.arange(-5, -4, h)
plt.plot(x, F(x))
plt.hlines(0, -5, -4)
plt.vlines(x_sol, -4, 4, colors = "green")
plt.show()
print("График F(x) на отрезке -5..-4 и решение")
```

```python
def newtone_method(eps):
  x_cur = -5
  k = 0
  while np.abs(F(x_cur) / d(F, x_cur)) > eps:
    x_cur = x_cur - F(x_cur) / d(F, x_cur)
    k += 1
  return k, x_cur

k, x_sol = newtone_method(eps)
print("Метод Ньютона")
print("Eps =", eps)
print("Число итераций", k)
print("Уточнённое решение", x_sol)
```
